```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib.gridspec import GridSpec
import numpy as np
from matplotlib import rcParams
```

```python
import os
# Get the current working directory
current_dir = os.getcwd()
# Set the data directory to the parent directory of the current working directory
wdir = os.path.dirname(current_dir)
# Change the current working directory to the parent directory
os.chdir(wdir)
```

```python
# Set the font globally
rcParams['font.family'] = 'Arial'
rcParams['font.size'] = 9
rcParams['axes.titlesize'] = 9
rcParams['axes.labelsize'] = 9
rcParams['xtick.labelsize'] = 9
rcParams['ytick.labelsize'] = 9
rcParams['legend.fontsize'] = 9
```

```python
# Load data
genetic_diversity_1w_mu_1 = np.loadtxt("data/Fig_2/genetic_diversity_one_week_mt_rate_0.1.txt")
genetic_diversity_1w_mu_2 = np.loadtxt("data/Fig_2/genetic_diversity_one_week_mt_rate_0.01.txt")
genetic_diversity_1w_mu_3 = np.loadtxt("data/Fig_2/genetic_diversity_one_week_mt_rate_0.001.txt")

genetic_diversity_1m_mu_1 = np.loadtxt("data/Fig_2/genetic_diversity_one_month_mt_rate_0.1.txt")
genetic_diversity_1m_mu_2 = np.loadtxt("data/Fig_2/genetic_diversity_one_month_mt_rate_0.01.txt")
genetic_diversity_1m_mu_3 = np.loadtxt("data/Fig_2/genetic_diversity_one_month_mt_rate_0.001.txt");
```

```python
def cm2inch(x): return float(x)/2.54
plotwidth    = cm2inch(8.90)
```

```python
base_=9
# Define the figure and axes
fig, axs = plt.subplots(1, 2, figsize=(1.5*plotwidth, plotwidth))  # Adjust the figsize according to your layout

# Plot for one week
axs[0].plot(np.arange(11), genetic_diversity_1w_mu_1, color=(0.5, 0, 0.5), linewidth=2, label=r'$\mu=10^{-1}$')
axs[0].plot(np.arange(11), genetic_diversity_1w_mu_2, color=(0.9, 0.5, 0.9), linewidth=2, label=r'$\mu=10^{-2}$')
axs[0].plot(np.arange(11), genetic_diversity_1w_mu_3, color=(0.5, 0.5, 0.5), linewidth=2, label=r'$\mu=10^{-3}$')
axs[0].spines['top'].set_visible(False)
axs[0].spines['right'].set_visible(False)

axs[0].set_xlim([0, 10])
axs[0].set_ylim([0, 1.01])
axs[0].set_xticks(np.arange(0, 11, 2))
axs[0].set_yticks([0, 1])
axs[0].set_xlabel("Number of mutations (α)")
axs[0].set_ylabel("Fraction of variants \n with more than α mutations")
axs[0].set_title('a  Generation time: one week', loc='left', fontname='Arial', fontsize=base_,pad=10,weight='bold')

axs[0].legend(loc='lower right',fontsize=base_,frameon=False)
axs[0].tick_params(axis='both', which='major',labelsize=base_)
# Plot for one month
axs[1].plot(np.arange(11), genetic_diversity_1m_mu_3, color=(0.6, 0.6, 0.6), linewidth=2, label=r'$\mu_N=\mu_B=10^{-1}$')
axs[1].plot(np.arange(11), genetic_diversity_1m_mu_2, color=(0.9, 0.5, 0.9), linewidth=2, label=r'$\mu_N=\mu_B=10^{-2}$')
axs[1].plot(np.arange(11), genetic_diversity_1m_mu_1, color=(0.5, 0, 0.5), linewidth=2, label=r'$\mu_N=\mu_B=10^{-3}$')

axs[1].spines['top'].set_visible(False)
axs[1].spines['right'].set_visible(False)
axs[1].spines['left'].set_visible(False)

axs[1].set_xlim([0, 10])
axs[1].set_ylim([0, 1.01])
axs[1].set_xticks(np.arange(0, 11, 2))
axs[1].set_yticks([])
axs[1].set_xlabel("Number of mutations (α)")
axs[1].set_ylabel("")
axs[1].set_title('b  Generation time: one month', loc='left', fontname='Arial',fontsize=base_,pad=10,weight='bold')

axs[1].legend(loc='upper right').set_visible(False)
axs[1].tick_params(axis='both', which='major')# THEindicated as defined.

#plt.subplots_adjust(left=0.01, right=0.6, top=0.99, bottom=0.01, wspace=0.2, hspace=-0.1)

#plt.subplots_adjust( bottom=0.15,top=0.8,hspace=0.2,wspace=0.08)
# Adjust layout
#fig.suptitle("Generation time: one week (a) Generation time: one month (b)",  fontsize=base_, x=0.4, y=0.95)
#fig.tight_layout()
#fig.subplots_adjust(bottom=0.4, left=0.001, right=0.99, top=0.9)

# Save the plot
#plt.savefig("/path/to/save/plot.pdf")
#plt.subplots_adjust( bottom=0.01,top=0.8,hspace=0.2,wspace=0.08)
#plt.subplots_adjust( bottom=0.15,top=0.8,hspace=0.2,wspace=0.08)
fig.tight_layout()
plt.subplots_adjust(bottom=0.1,hspace=0.2,wspace=0.08)
plt.show()
```
