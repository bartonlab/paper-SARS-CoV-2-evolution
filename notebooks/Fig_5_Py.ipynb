```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
import os
# Get the current working directory
current_dir = os.getcwd()
# Set the data directory to the parent directory of the current working directory
wdir = os.path.dirname(current_dir)
# Change the current working directory to the parent directory
os.chdir(wdir)
```

```python
# Set the font globally
rcParams['font.family'] = 'Arial'
rcParams['font.size'] = 9
rcParams['axes.titlesize'] = 9
rcParams['axes.labelsize'] = 9
rcParams['xtick.labelsize'] = 9
rcParams['ytick.labelsize'] = 9
rcParams['legend.fontsize'] = 9
```

```python
# Load data
# Parameters
T = 1000
fraction_mut_null = np.loadtxt("data/Fig_5/average_number_of_mutations_no_cd_bf_m_0.001_t2_2.txt")
fraction_mut_no_jump = np.loadtxt("data/Fig_5/average_number_of_mutations_cd_0.0004_bf_m_0.001_t2_2_no_jump.txt")
fraction_mut_cd = np.loadtxt("data/Fig_5/average_number_of_mutations_single_inidiv_0.001.txt")
fraction_mut_jump = np.loadtxt("data/Fig_5/average_number_of_mutations_cd_0.0004_bf_m_0.001_t2_2_jump.txt")
# Maximum slopes distribution
max_slopes_null = np.loadtxt("data/Fig_5/max_slope_null_bf_m_0.001_t2_2.txt")
max_slopes_no_jump = np.loadtxt("data/Fig_5/max_slope_no_jump_cd_0.0004_bf_m_0.001_t2_2.txt")
max_slopes_jump = np.loadtxt("data/Fig_5/max_slope_jump_cd_0.0004_bf_m_0.001_t2_2.txt");
```

```python
def cm2inch(x): return float(x)/2.54
plotwidth    = cm2inch(8.90)
```

```python
# Plot
fig, ax = plt.subplots(figsize=(1.5*plotwidth, plotwidth))

ax.plot(range(1, T + 2), fraction_mut_null, color='red', linewidth=2, label='Between Individuals - No Chronic Disease')
ax.plot(range(1, T + 2), fraction_mut_no_jump, color='green', linewidth=2, label='Between Individuals - Chronic Disease - No Mutation Burst')
ax.plot(range(1, T + 2), fraction_mut_jump, color='cyan', linewidth=2, label='Between Individuals - Chronic Disease - Mutation Burst')
ax.plot(range(1, T + 2), fraction_mut_cd, color='purple', linewidth=2, linestyle='--', label='Within Individual - Chronic Disease')

# Customizations
ax.set_yscale('log')
ax.set_ylim([0.001, 300])
ax.set_yticks([0.001, 0.01, 0.1, 1, 10, 100])
ax.legend(loc='lower right',frameon=False)
ax.grid(False)
ax.set_xlabel('Time (days)')
ax.set_ylabel('Average number of mutations per infection')
ax.set_title('')

# Adjust tick font size
ax.tick_params(axis='both', which='major')
#fig.subplots_adjust(bottom=0.4, left=0.001, right=0.99, top=0.9)
#plt.subplots_adjust( bottom=0.15,top=0.8,hspace=0.2,wspace=0.08)
#plt.subplots_adjust( bottom=0.1,top=0.8,hspace=0.2,wspace=0.08)
fig.tight_layout()
plt.subplots_adjust(bottom=0.1,hspace=0.2,wspace=0.08)
plt.show()
```
