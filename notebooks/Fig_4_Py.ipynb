```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
from matplotlib import rcParams
```

```python
import os
# Get the current working directory
current_dir = os.getcwd()
# Set the data directory to the parent directory of the current working directory
wdir = os.path.dirname(current_dir)
# Change the current working directory to the parent directory
os.chdir(wdir)
```

```python
# Set the font globally
rcParams['font.family'] = 'Arial'
rcParams['font.size'] = 9
rcParams['axes.titlesize'] = 9
rcParams['axes.labelsize'] = 9
rcParams['xtick.labelsize'] = 9
rcParams['ytick.labelsize'] = 9
rcParams['legend.fontsize'] = 9
```

```python
# Load data
data = np.loadtxt("data/Fig_4/heatmap_1.txt")
y = np.loadtxt("data/Fig_4/pc_heatmap_1.txt")
t_ = np.loadtxt("data/Fig_4/times_heatmap_1.txt");
```

```python
def cm2inch(x): return float(x)/2.54

plotwidth    = cm2inch(8.90)
```

```python

# Create the heatmap
#fig=plt.figure(figsize=(6, 4))
fig=plt.figure(figsize=(1.5*plotwidth, plotwidth))

heatmap = plt.imshow(data, aspect='auto', cmap='Blues',origin='lower',
                     extent=[t_.min()-0.5, t_.max()+0.5, np.log10(y).min(), np.log10(y).max()],vmin=0.0, vmax=0.12)
cbar = plt.colorbar(heatmap,  pad=0.02)
cbar.set_label('Number of mutational bursts per chronic disease cases', color='black')

#plt.colorbar(heatmap)
#plt.colorbar(label='Number of mutational bursts per chronic disease cases')
#cbar.ax.tick_params(labelsize=7,direction='inout')  # Set color bar tick label size and direction

cbar.ax.yaxis.set_tick_params(direction='inout', labelcolor='black')  # Example customization
#cbar.ax.tick_params(labelsize=7, direction='inout', fontname='Arial')  # Set color bar tick label size, direction, and font family
# Set font family for color bar tick labels
for tick_label in cbar.ax.yaxis.get_ticklabels():
    tick_label.set_fontname('Arial')
#plt.colorbar(label='Number of mutational bursts per chronic disease cases')

#cbar.set_label('Number of mutational bursts per chronic disease cases')

# Add titles and labels
plt.title("Mutation rates: $\mu_N=10^{-4},\mu_B=10^{-2}$")
plt.xlabel("Acute generation time ($t_a$)")
plt.ylabel("Fraction of chronic infections ($\log_{10}(p_c)$)")
# Invert y-axis if needed
#plt.gca().invert_yaxis()
#plt.subplots_adjust( bottom=0.15,top=0.8,hspace=0.2,wspace=0.08)
#plt.subplots_adjust( bottom=0.01,top=0.8,hspace=0.2,wspace=0.08)
#plt.subplots_adjust( bottom=0.1,top=0.7,hspace=0.2,wspace=0.08)
#plt.subplots_adjust(bottom=0.01,top=0.8,hspace=0.05,wspace=0.05)
fig.tight_layout()
plt.subplots_adjust(bottom=0.1,hspace=0.2,wspace=0.08)
plt.show()
```
