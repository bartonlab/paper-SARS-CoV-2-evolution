```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib.gridspec import GridSpec
from matplotlib import rcParams
```

```python
import os
# Get the current working directory
current_dir = os.getcwd()
# Set the data directory to the parent directory of the current working directory
wdir = os.path.dirname(current_dir)
# Change the current working directory to the parent directory
os.chdir(wdir)
```

```python
# Set the font globally
rcParams['font.family'] = 'Arial'
rcParams['font.size'] = 6
rcParams['axes.titlesize'] = 6
rcParams['axes.labelsize'] = 6
rcParams['xtick.labelsize'] = 6
rcParams['ytick.labelsize'] = 6
rcParams['legend.fontsize'] = 6
```

```python
def cm2inch(x): return float(x)/2.54
plotwidth    = cm2inch(8.90)
```

```python
# Load subplots 
diag0 = mpimg.imread("data/Fig_1/Fig_1_a_5.png")
diag1 = mpimg.imread("data/Fig_1/Transmission_3.png")
diag2 = mpimg.imread("data/Fig_1/Intrahost_mut_trajectories_7.png")
diag3 = mpimg.imread("data/Fig_1/fan_tree_2.png")


# Define the size of the layout
fig = plt.figure(figsize=(3*plotwidth, plotwidth))

# Define a 2x2 grid layout
gs = GridSpec(2, 2, figure=fig, height_ratios=[0.4, 0.7], width_ratios=[0.5, 0.5])

# Create subplots
ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[0, 1])
ax3 = fig.add_subplot(gs[1, 0])
ax4 = fig.add_subplot(gs[1, 1])
# Display images in the subplots
ax1.imshow(diag0)
ax2.imshow(diag2)
ax3.imshow(diag1)
ax4.imshow(diag3)

# Remove axes for each subplot
for ax in [ax1, ax2, ax3, ax4]:
    ax.axis('off')
# Set titles for each subplot
base_=7
ax1.set_title('a  Within host evolution', loc='left', fontname='Arial',pad=20,weight='bold')
ax2.set_title('b Within host mutation trajectories', loc='left',  fontname='Arial',pad=14,weight='bold')
ax3.set_title('c Acute and chronic cases give rise to distinct transmission chains', loc='left',  fontname='Arial',weight='bold')
ax4.set_title('d Transmission phylogeny', loc='left',  fontname='Arial', weight='bold')

plt.subplots_adjust(left=0.01, right=0.6, top=0.99, bottom=0.01, wspace=0.2, hspace=-0.1)
plt.show()
```
